#### Marital Status
Married-civ-spouse, Divorced, Never-married, Separated, Widowed, Married-spouse-absent, Married-AF-spouse.


```python
import numpy as np
import pandas as pd

data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None)
data.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary-class']
nation = data['native-country']
nationCount = nation.value_counts()
married = data['marital-status']
marriedCount = married.value_counts()
```

# Plotting Marital Status

```python
import matplotlib.pyplot as plt
#Data to plot
labels = 'Married-civ-spouse', 'Divorced', 'Never-married', 'Separated', 'Widowed', 'Married-spouse-absent', 'Married-AF-spouse'
sizes1 = [marriedCount[0],marriedCount[1],marriedCount[2],marriedCount[3],marriedCount[4],marriedCount[5],
          marriedCount[6]]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','red']
explode = (0.05, 0.05, 0.15, 0.45, 0.45, 0.45, 0.45)  # explode 1st slice
 
# Plot over 50K
#plt.pie(sizes1, explode=explode, labels=labels, colors=colors,
#        autopct='%1.1f%%', shadow=True, startangle=200)
plt.pie(sizes1, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Marital Status')
plt.axis('equal')
plt.show()
```
