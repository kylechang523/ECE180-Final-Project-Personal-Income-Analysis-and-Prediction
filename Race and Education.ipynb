```python
# Race
# White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black. 
whiteover = 0
whiteund = 0
APIover = 0
APIund = 0
AIEover = 0
AIEund = 0
blackover = 0
blackund = 0
otherover = 0
otherund = 0

#Education
#Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 
#9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool






above = []
below = []



linenum = 0
with open("adult.data.txt") as f:
    for line in f:
        #print (line)
        if line.count('White') == 1 and line.count('>50K') == 1:
            whiteover = whiteover + 1
        if line.count('White') == 1 and line.count('<=50K') == 1:
            whiteund = whiteund + 1
        if line.count('Asian-Pac-Islander') == 1 and line.count('>50K') == 1:
            APIover = APIover + 1
        if line.count('Asian-Pac-Islander') == 1 and line.count('<=50K') == 1:
            APIund = APIund + 1
        if line.count('Amer-Indian-Eskimo') == 1 and line.count('>50K') == 1:
            AIEover = AIEover + 1
        if line.count('Amer-Indian-Eskimo') == 1 and line.count('<=50K') == 1:
            AIEund = AIEund + 1
        if line.count('Black') == 1 and line.count('>50K') == 1:
            blackover = blackover + 1
        if line.count('Black') == 1 and line.count('<=50K') == 1:
            blackund = blackund + 1
        if line.count('Other') == 1 and line.count('>50K') == 1:
            otherover = otherover + 1
        if line.count('Other') == 1 and line.count('<=50K') == 1:
            otherund = otherund + 1
        
        
        linenum = linenum + 1
        #line = f.readline()
        if '<=50K' in line:
            below.append(line)
        elif '>50K' in line:
            above.append(line)
        
    f.close()

    
print (whiteover)
print (whiteund)
print (APIover)
print (APIund)
print (AIEover)
print (AIEund)
print (blackover)
print (blackund)
print (otherover)
print (otherund)
print(whiteover+whiteund+APIover+APIund+AIEover+AIEund+blackover+blackund+otherover+otherund)

print (linenum)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None)
data.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary-class']

over = data[data['salary-class']==' >50K']
und = data[data['salary-class']==' <=50K']

eduover = over['education'].value_counts()
raceover = over['race'].value_counts()

eduund = und['education'].value_counts()
raceund = und['race'].value_counts()


eduoversize = []
eduundsize = []
raceoversize = []
raceundsize = []

for i in range(len(eduover)):
    eduoversize.append(eduover[i])

for i in range(len(eduund)):
    eduundsize.append(eduund[i])
    
for i in range(len(raceover)):
    raceoversize.append(raceover[i])
    
for i in range(len(raceund)):
    raceundsize.append(raceund[i])
    
    
# Data to plot 
# Plot over 50K for edu
explode = np.zeros(len(eduover))+0.1
eduoverlabel = list(eduover.index)
plt.pie(eduoversize,explode = explode,labels=eduoverlabel,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income over 50K Dollars for different education level')
plt.axis('equal')
plt.show()

#Plor under 50K for edu
explode = np.zeros(len(eduund))+0.1
eduundlabel = list(eduund.index)
plt.pie(eduundsize,explode = explode,labels=eduundlabel,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income under 50K Dollars for different education level')
plt.axis('equal')
plt.show()

#Plor over 50K for race
explode = np.zeros(len(raceover))+0.1
raceoverlabel = list(raceover.index)
plt.pie(raceoversize,explode = explode,labels=raceoverlabel,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income under 50K Dollars for race')
plt.axis('equal')
plt.show()

#Plor under 50K for race
explode = np.zeros(len(raceund))+0.1
raceundlabel = list(raceund.index)
plt.pie(raceundsize,explode = explode,labels=raceundlabel,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income under 50K Dollars for race')
plt.axis('equal')
plt.show()
```
