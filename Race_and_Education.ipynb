```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches

data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None)
data.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary-class']

over = data[data['salary-class']==' >50K']
und = data[data['salary-class']==' <=50K']

eduover = over['education'].value_counts()
raceover = over['race'].value_counts()

eduund = und['education'].value_counts()
raceund = und['race'].value_counts()


eduoversize = []
eduundsize = []
raceoversize = []
raceundsize = []

for i in range(len(eduover)):
    eduoversize.append(eduover[i])

for i in range(len(eduund)):
    eduundsize.append(eduund[i])
    
for i in range(len(raceover)):
    raceoversize.append(raceover[i])
    
for i in range(len(raceund)):
    raceundsize.append(raceund[i])
    
    
# Data to plot 
# Plot over 50K for edu
explode = np.zeros(len(eduover))+0.1
eduoverlabel = list(eduover.index)
fig = plt.figure(figsize=[15, 15])
patches ,texts, autotexts= plt.pie(eduoversize,explode = explode,labels=None,
        autopct='%1.1f%%', shadow=True, startangle=50)

plt.title('Income over 50K Dollars for different education level')
plt.axis('equal')
plt.legend(patches, eduoverlabel, loc='upper left', bbox_to_anchor=(-0.1, 1.),
           fontsize=8)

plt.show()












#Plor under 50K for edu
explode = np.zeros(len(eduund))+0.1
eduundlabel = list(eduund.index)
plt.pie(eduundsize,explode = explode,labels=eduundlabel,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income under 50K Dollars for different education level')
plt.axis('equal')
plt.show()

#Plor over 50K for race
explode = np.zeros(len(raceover))+0.1
raceoverlabel = list(raceover.index)
plt.pie(raceoversize,explode = explode,labels=raceoverlabel,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income under 50K Dollars for race')
plt.axis('equal')
plt.show()

#Plor under 50K for race
explode = np.zeros(len(raceund))+0.1
raceundlabel = list(raceund.index)
plt.pie(raceundsize,explode = explode,labels=raceundlabel,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income under 50K Dollars for race')
plt.axis('equal')
plt.show()
```

```python
import string
import seaborn as sns
names =['age','workplace','fnlwgt','education','working-num','marital-status',
           'occupation','relationship','race','sex','capital-gain','capital-loss',
           'hours-per-week','native-country','income']
df_train = pd.read_csv('adult.data.txt',header=None,names=names)
```

```python
total = float(len(df_train['race']))

plt.figure(figsize=(15,10))
ax = sns.countplot(df_train['race'],order = df_train['race'].value_counts().iloc[:10].index,palette=sns.color_palette("husl", 10))
plt.title('Top 5 Race Categories', fontsize = 15)
plt.ylabel('Count of item', fontsize = 10)
plt.xlabel('Race')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:.2f}%'.format((height/total)*100),
            ha="center") 
plt.savefig('race.png', bbox_inches='tight')
plt.show()
```
