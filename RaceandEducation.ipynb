```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches

data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None)
data.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary-class']

over = data[data['salary-class']==' >50K']
und = data[data['salary-class']==' <=50K']

eduover = over['education'].value_counts()
raceover = over['race'].value_counts()

eduund = und['education'].value_counts()
raceund = und['race'].value_counts()


eduoversize = []
eduundsize = []
raceoversize = []
raceundsize = []

for i in range(len(eduover)):
    eduoversize.append(eduover[i])

for i in range(len(eduund)):
    eduundsize.append(eduund[i])
    
for i in range(len(raceover)):
    raceoversize.append(raceover[i])
    
for i in range(len(raceund)):
    raceundsize.append(raceund[i])
    
    
# Data to plot 
# Plot over 50K for edu

sizePercent = ["{0:.2f}".format(x/float(sum(eduoversize))*100) for x in eduoversize]

explode = np.zeros(len(eduover))+0.1
eduoverlabel = list(eduover.index)
fig = plt.figure(figsize=[5, 5])
patches ,texts= plt.pie(eduoversize,explode = explode,labels=None,
        shadow=True, startangle=50)

aaa=[None]*len(eduoverlabel)
for i in range(len(eduoverlabel)):
    aaa[i] = str(eduoverlabel[i])+ ' '+str(sizePercent[i]) + '%'
plt.title('Income over 50K Dollars for different education level')
plt.axis('equal')
plt.legend(aaa, loc='upper right', bbox_to_anchor=(-.1, 1.),
           fontsize=8, prop={'size': 12})
plt.show()
#fig.savefig("Edu_over_50K.png")




#Plor under 50K for edu
sizePercent = ["{0:.2f}".format(x/float(sum(eduundsize))*100) for x in eduundsize]

explode = np.zeros(len(eduund))+0.1
eduundlabel = list(eduund.index)
fig = plt.figure(figsize=[5, 5])
plt.pie(eduundsize,explode = explode,labels=None,
        shadow=True, startangle=50)

aaa=[None]*len(eduundlabel)
for i in range(len(eduundlabel)):
    aaa[i] = str(eduundlabel[i])+ ' '+str(sizePercent[i]) + '%'
plt.title('Income under 50K Dollars for different education level')
plt.axis('equal')
plt.legend(aaa, loc='upper right', bbox_to_anchor=(-.1, 1.),
           fontsize=8, prop={'size': 12})
plt.show()




#Plor over 50K for race
sizePercent = ["{0:.2f}".format(x/float(sum(raceoversize))*100) for x in raceoversize]

explode = np.zeros(len(raceover))+0.1
raceoverlabel = list(raceover.index)
fig = plt.figure(figsize=[5, 5])
plt.pie(raceoversize,explode = explode,labels=None,
        shadow=True, startangle=50)

aaa=[None]*len(raceoverlabel)
for i in range(len(raceoverlabel)):
    aaa[i] = str(raceoverlabel[i])+ ' '+str(sizePercent[i]) + '%'
plt.title('Income over 50K Dollars for different race')
plt.axis('equal')
plt.legend(aaa, loc='upper right', bbox_to_anchor=(-.1, 1.),
           fontsize=8, prop={'size': 12})
plt.show()





#Plor under 50K for race
sizePercent = ["{0:.2f}".format(x/float(sum(raceundsize))*100) for x in raceundsize]

explode = np.zeros(len(raceund))+0.1
raceundlabel = list(raceund.index)
fig = plt.figure(figsize=[5, 5])
plt.pie(raceundsize,explode = explode,labels=None,
        shadow=True, startangle=50)

aaa=[None]*len(raceundlabel)
for i in range(len(raceundlabel)):
    aaa[i] = str(raceundlabel[i])+ ' '+str(sizePercent[i]) + '%'
plt.title('Income under 50K Dollars for different race')
plt.axis('equal')
plt.legend(aaa, bbox_to_anchor=(-.1, 1.),loc='upper right',
           fontsize=1, prop={'size': 12})
plt.show()
#fig.savefig("Race_under_50K.png")
```

```python
import string
import seaborn as sns
names =['age','workplace','fnlwgt','education','working-num','marital-status',
           'occupation','relationship','race','sex','capital-gain','capital-loss',
           'hours-per-week','native-country','income']
df_train = pd.read_csv('adult.data.txt',header=None,names=names)
```

```python
total = float(len(df_train['race']))

plt.figure(figsize=(15,10))
ax = sns.countplot(df_train['race'],order = df_train['race'].value_counts().iloc[:10].index,palette=sns.color_palette("husl", 10))
plt.title('Top 5 Race Categories', fontsize = 15)
plt.ylabel('Count of item', fontsize = 10)
plt.xlabel('Race')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:.2f}%'.format((height/total)*100),
            ha="center") 
plt.savefig('race.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x = df_train.education,order = df_train.education.value_counts().iloc[:10].index, orient = 'v',palette=sns.color_palette("husl", 10))
plt.title('Top 10 Education level', fontsize = 20)
plt.ylabel('Category Name', fontsize = 15)
plt.xlabel('Count of Item in the Category', fontsize = 15)
plt.savefig('top_10_edu.png', bbox_inches='tight')
plt.show()
```
