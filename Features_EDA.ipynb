```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import string
import seaborn as sns
```

```python
names =['age','workplace','fnlwgt','education','working-num','marital-status',
           'occupation','relationship','race','sex','capital-gain','capital-loss',
           'hours-per-week','native-country','income']
df_train = pd.read_csv('adult.data.txt',header=None,names=names)
```

```python
df_train.age.describe()
```

```python
plt.figure(figsize=(15, 10))
plt.hist(df_train['age'], bins=20, range=[0,90], label='Age')
plt.title('Age Distribution', fontsize=20)
plt.xlabel('Age', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.legend(fontsize=15)
#plt.savefig('price_distri.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x=df_train.income, y = df_train.age, showfliers=False, orient = 'v',palette=sns.color_palette("husl", 10))
plt.title('The Effects of age on income', fontsize = 15)
plt.xlabel('Income Level', fontsize = 10)
plt.ylabel('age w/o outliers', fontsize = 10)
#plt.savefig('shipping_price.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y = df_train.education, \
              order = df_train.education.value_counts().iloc[:10].index, \
                                                      orient = 'h',palette=sns.color_palette("husl", 10))
plt.title('Top 10 Eudation', fontsize = 20)
plt.ylabel('Category Name', fontsize = 15)
plt.xlabel('Count of Item in the Category', fontsize = 15)
plt.savefig('top_10_cat.png', bbox_inches='tight')
plt.show()
```

```python
total = float(len(df_train.income))

plt.figure(figsize=(10,10))
ax = sns.countplot(df_train.income,palette=sns.color_palette("husl", 10))
plt.title('Income Category', fontsize = 15)
plt.ylabel('Count of item', fontsize = 10)
plt.xlabel('Income level')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:.2f}%'.format((height/total)*100),
            ha="center") 
#plt.savefig('shipping.png', bbox_inches='tight')
plt.show()
```
