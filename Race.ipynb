```python
# Race
# White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black. 
whiteover = 0
whiteund = 0
APIover = 0
APIund = 0
AIEover = 0
AIEund = 0
blackover = 0
blackund = 0
otherover = 0
otherund = 0

#Education
#Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 
#9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool

above = []
below = []

linenum = 0
with open("adult.data.txt") as f:
    for line in f:
        #print (line)
        if line.count('White') == 1 and line.count('>50K') == 1:
            whiteover = whiteover + 1
        if line.count('White') == 1 and line.count('<=50K') == 1:
            whiteund = whiteund + 1
        if line.count('Asian-Pac-Islander') == 1 and line.count('>50K') == 1:
            APIover = APIover + 1
        if line.count('Asian-Pac-Islander') == 1 and line.count('<=50K') == 1:
            APIund = APIund + 1
        if line.count('Amer-Indian-Eskimo') == 1 and line.count('>50K') == 1:
            AIEover = AIEover + 1
        if line.count('Amer-Indian-Eskimo') == 1 and line.count('<=50K') == 1:
            AIEund = AIEund + 1
        if line.count('Black') == 1 and line.count('>50K') == 1:
            blackover = blackover + 1
        if line.count('Black') == 1 and line.count('<=50K') == 1:
            blackund = blackund + 1
        if line.count('Other') == 1 and line.count('>50K') == 1:
            otherover = otherover + 1
        if line.count('Other') == 1 and line.count('<=50K') == 1:
            otherund = otherund + 1
        
        
        linenum = linenum + 1
        #line = f.readline()
        if '<=50K' in line:
            below.append(line)
        elif '>50K' in line:
            above.append(line)
        
    f.close()
 
print (whiteover)
print (whiteund)
print (APIover)
print (APIund)
print (AIEover)
print (AIEund)
print (blackover)
print (blackund)
print (otherover)
print (otherund)
print(whiteover+whiteund+APIover+APIund+AIEover+AIEund+blackover+blackund+otherover+otherund)

print (linenum)
```

```python
# White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black. 
import matplotlib.pyplot as plt
 
# Data to plot
labels = 'White', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Black', 'Other'
sizes1 = [whiteover, APIover, AIEover, blackover, otherover]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','red']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice
 
# Plot over 50K
plt.pie(sizes1, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=200)

plt.title('Income over 50K Dollars')
plt.axis('equal')
plt.show()


sizes2 = [whiteund, APIund, AIEund, blackund, otherund]

# Plot under 50K
plt.pie(sizes2, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=250)

plt.title('Income under 50K Dollars')
plt.axis('equal')
plt.show()

```
